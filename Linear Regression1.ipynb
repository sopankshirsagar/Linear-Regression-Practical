## Lineare Regression moled on california_housing dataset

```python
#importing dataset and linear miodel for linear regression 
from sklearn import linear_model,datasets
#importing metrics for measuring performence of model 
from sklearn.metrics import mean_squared_error,r2_score
```

```python
#Numpy for mathmatical methods and matplotlub and seaborn for Data visualization 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
#loading Dataset 
data_x,data_y = datasets.fetch_california_housing(return_X_y=True)
```

```python
#use only one feature 
data_x = data_x[:,np.newaxis,0]
```

```python
#splitting data_x into 80 % train and 20 % test data
print(20*data_x.shape[0]/100) #20 percent rows 
data_x_train = data_x[:-4128]
data_x_test =data_x[-4128:]
```

```python
#splitting data_y into train test
data_y_train = data_y[:-4128]
data_y_test = data_y[-4128:]
```

```python
# create linear regression model 
regr = linear_model.LinearRegression()
```

```python
# training model using training set 
regr.fit(data_x_train,data_y_train)
```

```python
# making predictions using testing sets 
predict_data =regr.predict(data_x_test)
```

```python
#mean squared error
print("mean squared error :%.2f"%mean_squared_error(data_y_test,predict_data))
# r2 score coefficient of determination 
print("coefficient of determination :%.2f"%r2_score(data_y_test,predict_data))
```

```python
#plotting graph 
sns.scatterplot(x=data_y_test,y=predict_data,color='maroon')
plt.xticks()
plt.yticks()
plt.show()
```

```python

sns.histplot(data_y_test,kde=True,color='gold')
sns.histplot(predict_data,kde=True,color='black')
plt.title("Diff between Actual and predicted data")
```

```python
#here we cann see rellation od actual and predicted values 
sns.jointplot(x=data_y_test,y=predict_data,color='skyblue')
plt.xlabel("Actual data")
plt.ylabel("predicted data")
plt.show()
```
